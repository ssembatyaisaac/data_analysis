Dropping all excess columns that were empty and replacing 'NaN' cells with empty strings ''

```python
df.columns.str.match('Unnamed')
df2 = df.loc[:,~df.columns.str.match('Unnamed')]
df2.fillna('',inplace=True)
df2
```

Cleaning dates and converting to proper date time format

```python
# df2[['Date of Birth','P.L.E Year','U.C.E Year','U.A.C.E Year','Year of first Degree','Year of second degree','Year of Third Degree']]
# df2.loc[df2['Date of Birth'] == '07/071983']
# df2.loc[df2['Date of Birth'] == '09/0/1971']
df2['Date of Birth'] = df2['Date of Birth'].replace(['07/071983'],'07/07/1983')
df2['Date of Birth'] = df2['Date of Birth'].replace(['09/0/1971'],'09/01/1971')
# replace those dates with a proper date to avoid errors
df2['Date of Birth'] = pd.to_datetime(df2['Date of Birth'])
df2[['Date of Birth','P.L.E Year','U.C.E Year','U.A.C.E Year','Year of first Degree','Year of second degree','Year of Third Degree']].head(10)
```

Changing columns which have values with uneven cases to capitalized eg **DATa analySiS** to **Data Analysis**
and stripping all trailing whitespace

```python
df2['District']= df2['District'].str.strip().str.capitalize()
df2['Political Party']= df2['Political Party'].str.strip().str.capitalize()
df2['Name of M.P']= df2['Name of M.P'].str.strip().str.capitalize()
df2['Constituency'] = df2['Constituency'].str.strip().str.capitalize()
df2['Committee_1'] = df2['Committee_1'].str.strip().str.title()
df2['Committee_2'] = df2['Committee_2'].str.strip().str.title()
df2['Committee_3'] = df2['Committee_3'].str.strip().str.title()
df2
```

This function filters out oddities in the Marital status column eg **row 108 had phone numbers**

```python
#This function filters out oddities in the Marital status column eg row 108 had phone numbers
def filter_rows_by_values(df, col, values):
    return df[df[col].isin(values) == True]
df2 = filter_rows_by_values(df2, 'Marital status',['Married','Single','Widowed','Divorced'])
```
