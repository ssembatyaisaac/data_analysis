Dropping all excess columns that were empty and replacing 'NaN' cells with empty strings ''

```python
df.columns.str.match('Unnamed')
df2 = df.loc[:,~df.columns.str.match('Unnamed')]
df2.fillna('',inplace=True)
df2
```

Cleaning dates and converting to proper date time format

```python
# df2[['Date of Birth','P.L.E Year','U.C.E Year','U.A.C.E Year','Year of first Degree','Year of second degree','Year of Third Degree']]
# df2.loc[df2['Date of Birth'] == '07/071983']
# df2.loc[df2['Date of Birth'] == '09/0/1971']
df2['Date of Birth'] = df2['Date of Birth'].replace(['07/071983'],'07/07/1983')
df2['Date of Birth'] = df2['Date of Birth'].replace(['09/0/1971'],'09/01/1971')
# replace those dates with a proper date to avoid errors
df2['Date of Birth'] = pd.to_datetime(df2['Date of Birth'])
df2[['Date of Birth','P.L.E Year','U.C.E Year','U.A.C.E Year','Year of first Degree','Year of second degree','Year of Third Degree']].head(10)
```
